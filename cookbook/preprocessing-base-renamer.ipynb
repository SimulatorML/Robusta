# Import

```python
# Basic
import pandas as pd
import numpy as np

import warnings

warnings.simplefilter('ignore')

# ML Toolkit
from robusta.preprocessing import ColumnRenamer

# Data
from catboost.datasets import titanic
```

# Data

```python
df, _ = titanic()
df.head()
```

# Renamer

## â€“

```python
renamer = ColumnRenamer()
renamer.fit_transform(df).head()
```

## Iterable

```python
renamer = ColumnRenamer(range(12))
renamer.fit_transform(df).head()
```

```python
renamer = ColumnRenamer(np.random.permutation(df.columns))
renamer.fit_transform(df).head()
```

## String

```python
renamer = ColumnRenamer('x')
renamer.fit_transform(df).head()
```

## Prefix / Suffix

```python
renamer = ColumnRenamer(prefix='base_')
renamer.fit_transform(df).head()
```

```python
renamer = ColumnRenamer(suffix='_new')
renamer.fit_transform(df).head()
```

```python
renamer = ColumnRenamer(prefix='t_', suffix='+1')
renamer.fit_transform(df).head()
```

```python
renamer = ColumnRenamer('svd', prefix='pro_')
renamer.fit_transform(df).head()
```

```python
features = list(map(lambda x: chr(x+ord('A')), range(12)))
renamer = ColumnRenamer(features, suffix='_new')
renamer.fit_transform(df).head()
```

## Callable

```python
renamer = ColumnRenamer(lambda x: x.upper(), suffix='_up')
renamer.fit_transform(df).head()
```

```python
renamer = ColumnRenamer(lambda x: x[:3])
renamer.fit_transform(df).head()
```
